```python
def extraction_time(txt):
  check =False
  data = ""
  for i in range(len(txt)):
    if txt[i]== ')':
      for j in range(i+1,i+20):
        data +=txt[j]
      return data
```

```python
df['Total_time']=df['Title'].apply(extraction_time)
```

```python
import pandas as pd

# Extract start and end dates
df[["Start_Date", "End_Date"]] = df["Title"].str.extract(
    r'([A-Z][a-z]{2} \d{4}) - ([A-Z][a-z]{2} \d{4})'
)

# Convert to datetime
df["Start_Date"] = pd.to_datetime(df["Start_Date"], format="%b %Y")
df["End_Date"] = pd.to_datetime(df["End_Date"], format="%b %Y")

# Calculate total months (including both start and end months)
df["Total_Months"] = (
    (df["End_Date"].dt.year - df["Start_Date"].dt.year) * 12
    + (df["End_Date"].dt.month - df["Start_Date"].dt.month)
    + 1
)

# Display the result
print(df[["Start_Date", "End_Date", "Total_Months"]].head())
```
